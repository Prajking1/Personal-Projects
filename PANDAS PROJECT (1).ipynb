**Analyze the Federal Aviation Authority (FAA) Dataset using Pandas**

```python
import pandas as pd
import numpy as np
data = pd.read_csv("C:/Users/Dell/OneDrive/Desktop/faa_ai_prelim.csv")
data["index_no."] = np.arange(len(data))
```

```python
#2. Clean the dataset and replace the fatal flag NaN with “No”
df=data.fillna('No')
df.head()
```

```python
#3. Find the aircraft types and their occurrences in the dataset 
print("DIFFERENT AIRCRAFT TYPES")
print(df.ACFT_MAKE_NAME)
```

```python
#TO PRINT OCCURENCES OF AIRCRAFT MANUFACTURERS 

df["ACFT_MAKE_NAME"].value_counts().sort_index().plot.bar(width=0.5)
#value_counts gives total occurences
#sort_index sort them in ascending order
#plot.barh gives horizontal bar chart
```

```python
#4. Remove all the observations where aircraft names are not available 
df["ACFT_MAKE_NAME"].replace('', np.nan, inplace = True)

for items in df['ACFT_MAKE_NAME']:
    print(items)
```

```python
#As we observe above blank entries is replaced by NaN VALUES 
#Now we drop NaN VALUES to completely remove the entries

df.dropna(subset=['ACFT_MAKE_NAME'], inplace = True)
for items in data['ACFT_MAKE_NAME']:
    print(items)
```
