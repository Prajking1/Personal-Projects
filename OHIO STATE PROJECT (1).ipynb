**MAKING MORTALITY PIE CHART FOR EACH TYPE IN STATE OHIO FOR EACH YEAR**

```python
#MAKING PIE CHART FOR EACH MORTALITY DATASETS

df1 = df1.loc[df1['STATE']== 'OH']
df1 = df1.drop(['STATE'],axis=1)
```

```python
df1.head()
```

```python
df2 = df2.loc[df2['STATE']== 'OH']
df2 = df2.drop(['STATE'],axis=1)
```

```python
df3 = df3.loc[df3['STATE']== 'OH']
df3 = df3.drop(['STATE'],axis=1)
```

```python
df4 = df4.loc[df4['STATE']== 'OH']
df4 = df4.drop(['STATE'],axis=1)
```

```python
df4.head()
```

```python
data = pd.concat([df1 , df2 , df3 , df4], axis=1 , join='inner')
display(data)
```

```python
data.info()
```

```python
data.RATE = data.RATE.astype('float64')
```

```python
data.RATE.dtypes
```

```python
data1 = np.array(data)
np.set_printoptions(suppress=True)
print(data1)
```

```python
keys = ['Infant Mortality','Heart disease Mortality','Cancer Mortality','Accident Mortality']
```

```python
pal_colors = sns.color_palette('bright')
dt1 = [data1[0][2],data1[0][5],data1[0][8],data1[0][11]]
plt.title("2020 NUMBER OF DEATHS FROM VARIOUS CAUSES")
plt.pie(dt1,labels = keys,colors = pal_colors,autopct='%0f%%')
plt.show()
```
