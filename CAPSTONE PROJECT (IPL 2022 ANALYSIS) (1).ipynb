```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set()
```

```python
df = pd.read_csv('C:/Users/Dell/OneDrive/Desktop/ipl_22.csv')
df.head(10)
```

**MOST MATCH WINNING TEAM**

```python
sns.countplot(y= 'match_winner', data= df, palette ='magma' )
plt.title("MOST MATCH WINNING TEAM IN IPL 2022 ")
```

**MATCHES WON BY CHASING VS DEFENDING**

```python
d = df['won_by'].value_counts()
print(d)
```

```python
keys = ['CHASING','DEFENDING']
cp = sns.color_palette('bright')
plt.title('MATCHES WON BY CHASING VS DEFENDING')
plt.pie(d, labels=keys, colors=cp,autopct='%0f')
plt.show()                                      
```

**WHAT MOST TEAM PREFERED BATTING OR FIELDING**

```python
toss=df['toss_decision'].value_counts()
print(toss)
```

```python

sns.countplot(y='toss_winner' , hue='toss_decision', palette = 'rocket', data=df)
```

**TOP SCORERS IN IPL 2022**

```python
fig,ax = plt.subplots(figsize=(20,5))
d = df['top_scorer'].value_counts()
freq= d.index
val_u = d.values
ax.bar(freq, val_u, width=0.8 ,color=['yellow', 'purple'])

plt.xticks(rotation=60)
ax.set_title('TOP SCORERS IN IPL 2022')
ax.set_xlabel('PLAYER NAME')
ax.set_ylabel('NUMBER OF MATCHES')
```

**MOST PLAYER OF THE MATCH RECORDS**

```python
fig,ax = plt.subplots(figsize=(20,5))
d1 = df['player_of_the_match'].value_counts()
freq1= d1.index
val_u1 = d1.values
ax.bar(freq1, val_u1, width=0.8 ,color=['red', 'green'])

plt.xticks(rotation=75)
ax.set_title('MOST PLAYER OF THE MATCH RECORDS IN IPL 2022')
ax.set_xlabel('PLAYER NAME')
ax.set_ylabel('NUMBER OF MATCHES')
```

**BEST BOWLERS IN IPL 2022**

```python
fig,ax = plt.subplots(figsize=(20,5))

d2 = df["best_bowling"].value_counts()
freq2 = d2.index
val_u2 = d2.values

ax.bar(freq2, val_u2, width=0.8, color=['black','cyan'])
plt.xticks(rotation=60)

ax.set_title("BEST BOWLERS IN IPL 2022")
ax.set_xlabel('PLAYER NAME')
ax.set_ylabel("NUMBER OF MATCHES")
```
