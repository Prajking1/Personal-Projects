```python
df = df[df.horsepower != '?']
df.horsepower = df.horsepower.astype('float64')
df.dtypes
```

**FEATURE ENGINEERING ON DATASET**

```python
#ADDING POWER/WEIGHT (P/W) RATIO TO DATAFRAME
df['P/W RATIO'] = (df.horsepower/df.weight)
```

```python
#creating displacement in CC
df['DispCC'] = df['displacement']*16.3871
#converting CC into litres
df['Displtr'] = df['DispCC']/1000
```

```python
df.describe()
#descriptive stats on the dataframe
```

**EDA**

```python
sns.displot(df['mpg'], kde= True)
#using histogram
plt.title('MPG Distribution in data')
plt.show()
```

```python
sns.histplot(df['acceleration'], kde =True , color = 'red')
plt.title('Acceleration Distribution')
plt.show()
```
