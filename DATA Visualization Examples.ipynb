We can make graph more illustrative by coloring each data point by its class , we can do it by making color dict. mapping species to color code and then scattering each point using for loop and passing respective color




**MATPLOTLIB**


```python
#color dictionary
colors = {'Iris-setosa':'r','Iris-versicolor':'g','Iris-virginica':'b'}

fig,ax = plt.subplots()
#plot each datapoint
for i in range(len(iris['SepalLengthCm'])):
    ax.scatter(iris['SepalLengthCm'][i],iris['SepalWidthCm'][i],color =colors[iris['Species'][i]])

#set title 
ax.set_title('iris dataset')
ax.set_xlabel('SepalLengthCm')
ax.set_ylabel('SepalWidthCm')
```

**SEABORN**



```python
sns.scatterplot(x='SepalLengthCm',y='SepalWidthCm', hue='Species', data=iris)
```

**LINE CHART**

using plot method , using this we can plot multiple columns in one graph and axis by looping through each column

```python
#get columns dat we hv to plot
columns = iris.columns.drop(['Species','Id'])
fig,ax = plt.subplots()

#create x data for total iterable object range
x_data = range(0,iris.shape[0])

#plot each column
for column in columns:
    ax.plot(x_data,iris[column])
    
ax.set_title("Iris Dataset")
ax.legend(iris[columns])
```

**BAR CHART(in matplotlib)**

Bar chart can be plotted using bar method ,unlike histogram method it doesnt count freq automatically henceforth pandas value_counts function is used. bar chart is useful for data with less categories(less than 30) or else it gets messy

```python
fig,ax = plt.subplots()
#counting occurence of each class
data = wine_reviews['points'].value_counts()
#get x and y data
points = data.index
frequency = data.values
#create bar chart
ax.bar(points,frequency)
#set titles and labels
ax.set_title("Wine Review Score")
ax.set_xlabel('Points')
ax.set_ylabel('frequency')
```

**IN PANDAS**
